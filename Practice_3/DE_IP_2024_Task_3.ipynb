```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import copy
```

```python
img = cv2.imread('sar_3.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

plt.imshow(img)
plt.show()
```

```python
blur = cv2.GaussianBlur(gray, (5, 5), 0)
plt.imshow(blur, cmap='gray')
plt.show()
```

```python
edges = cv2.Canny(blur, 50, 150, apertureSize=3)

plt.imshow(edges, cmap='gray')
plt.show()
```

```python
lines = cv2.HoughLines(edges, 1, np.pi / 180, 120)

if lines is not None:
    best_pts = None
    max_len = 0.0
    for rho, theta in lines[:, 0]:
        a = np.cos(theta)
        b = np.sin(theta)
        x0, y0 = a * rho, b * rho
        pt1 = (int(x0 + 1000 * (-b)), int(y0 + 1000 * (a)))
        pt2 = (int(x0 - 1000 * (-b)), int(y0 - 1000 * (a)))
        length = np.hypot(pt1[0] - pt2[0], pt1[1] - pt2[1])
        if length > max_len:
            max_len = length
            best_pts = (pt1, pt2)
    if best_pts is not None:
        cv2.line(img, best_pts[0], best_pts[1], (255, 0, 0), 7, cv2.LINE_AA)

plt.imshow(img)
plt.show()
```

```python
img = cv2.imread('sar_3.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(gray, cmap='gray')
plt.show()
```

```python
binary = copy.deepcopy(gray)
threshold = 50

binary[gray < threshold] = 0
binary[gray >= threshold] = 255

plt.imshow(binary, cmap='gray')
plt.show()
```

```python
_, binary_otsu = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

plt.imshow(binary_otsu, cmap='gray')
plt.show()
```

```python
binary_adaptive = cv2.adaptiveThreshold(
    gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 71, 21
)

plt.imshow(binary_adaptive, cmap='gray')
plt.show()
```

```python
scale = 0.5
delta = 0.1
ddepth = cv2.CV_16S

grad_x = cv2.Sobel(gray, ddepth, 1, 0, ksize=3, scale=scale, delta=delta, borderType=cv2.BORDER_DEFAULT)
grad_y = cv2.Sobel(gray, ddepth, 0, 1, ksize=3, scale=scale, delta=delta, borderType=cv2.BORDER_DEFAULT)

plt.imshow((grad_x - grad_x.min()) * 255, cmap='gray')
plt.show()
```

```python
plt.imshow((grad_y - grad_y.min()) * 255, cmap='gray')
plt.show()
```

```python
grad = cv2.addWeighted(grad_x, 0.5, grad_y, 0.5, 0.0)

plt.imshow((grad - grad.min()) * 255, cmap='gray')
plt.show()
```
